```python
x=cdata[['Age','Fare']] #taking two indepedent variables
```

```python
y=cdata[["Survived"]] #taking dependent variables
```

```python
from sklearn.model_selection import train_test_split # spliting data into two parts train and test , test size =25%
xtrain, xtest, ytrain, ytest = train_test_split( x, y, test_size = 0.25)
```

```python
from sklearn.preprocessing import StandardScaler #compressing data into smaller parts for better output
sc_x = StandardScaler() #standardization
xtrain = sc_x.fit_transform(xtrain) 
xtest = sc_x.transform(xtest) 
sc_x
```

```python
from sklearn.linear_model import LogisticRegression #fitting our data into logical regression
classifier = LogisticRegression() 
classifier.fit(xtrain, ytrain) 
```

```python
y_pred = classifier.predict(xtest) # checking our test data
```

```python
from sklearn.metrics import confusion_matrix 
cm = confusion_matrix(ytest, y_pred) 
print ("Confusion Matrix : \n", cm) 
```

```python
from sklearn.metrics import accuracy_score #checking our model occuracy
print ("Accuracy : ", accuracy_score(ytest, y_pred)) 
```

```python
from matplotlib.colors import ListedColormap 
X_set, y_set = xtest, ytest 
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01), 
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01)) 

plt.contourf(X1, X2, classifier.predict( np.array([X1.ravel(), X2.ravel()]).T).reshape( X1.shape), alpha = 0.75, cmap = ListedColormap(('red', 'green'))) 

plt.xlim(X1.min(), X1.max()) 
plt.ylim(X2.min(), X2.max()) 

for i, j in enumerate(np.unique(y_set)): plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j) 
plt.title('Classifier (Test set)') 
plt.xlabel('Age') 
plt.ylabel('survived') 
plt.legend() 
plt.show() 
```
